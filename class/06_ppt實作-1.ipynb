```python
from bs4 import BeautifulSoup
import requests



url = 'https://www.ptt.cc/bbs/movie/index{}.html'

userAgent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'

headers = {'User-Agent' : userAgent}


page = 9506

for i in range(0,5):
    res = requests.get(url.format(page), headers=headers)

    soup = BeautifulSoup(res.text, 'html.parser') #設定轉換器 。 如果要轉lxml -> pip install lxml
    #print(soup)


    titleTagList = soup.select('div.title')
    #print(titleTagList)


    for titleTag in titleTagList:
       try:
        title = titleTag.select('a')[0].text 
        articleUrl = 'https://www.ptt.cc/' + titleTag.select('a')[0]['href']
        print(title)
        print(articleUrl)
       except IndexError as e :
        print(titleTag)
       print('==============')

    page -= 1
```
