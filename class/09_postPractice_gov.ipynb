```python
import requests
from bs4 import BeautifulSoup


url = 'https://web.pcc.gov.tw/tps/pss/tender.do?searchMode=common&searchType=advance'

userAgent = 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'

headers = {'User-Agent' : userAgent}


dataStr = """method: search
searchMethod: true
searchTarget: ATM
orgName: ergfve
orgId: bfzddf
hid_1: 1
tenderName: fdbzdfbdf
tenderId: vcb cvb cv
tenderStatus: 5,6,20,28
tenderWay: 
awardAnnounceStartDate: 110/05/13
awardAnnounceEndDate: 110/05/13
proctrgCate: 
tenderRange: 
minBudget: 
maxBudget: 
item: v vcv
hid_2: 1
gottenVendorName:  vcvc 
gottenVendorId: vc z bcvz
hid_3: 1
submitVendorName: c vc vc
submitVendorId: cv vc bvc
location: 
execLocationArea: 
priorityCate: 
isReConstruct: 
btnQuery: 查詢"""

data = dict()

#寫法1
for row in dataStr.split("\n"):
    print(row)
    data[row.split(': ')[0]] = row.split(": ")[1] 
    
    
#寫法2

data = {row.split(': ')[0] : row.split(": ")[1] for row in dataStr.split("\n")}  #面試必考

res = requests.post(url, headers=headers, data=data)
print(res.text)



```

```python
url = 'https://web.pcc.gov.tw/tps/pss/tender.do?method=goSearch&searchMode=common&searchType=advance&searchTarget=ATM'


res = requests.get(url, headers=headers)


soup = BeautifulSoup(res.text, 'html.parser')

soup.select('input[type="hidden"]') #先搜尋input標籤，但由於要進一步搜尋 type=hidden，因此對input使用[]，進一步搜尋


```

```python
data = dict()

for t in soup.select('input[type="hidden"]'):
    try:
        print('Key:', t['name'], 'Value:', t['value'] )
        data[t['name']] = t['value']
    except KeyError:
        pass
    
data
```
